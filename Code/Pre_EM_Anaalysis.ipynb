## Mapping 

```python
import pandas as pd
import numpy as np
from tqdm.notebook import tnrange

# Define chunk size
chunk_size = 10000  # Adjust based on your system's memory capacity

file_path=('C:\\Users\\rp7248\\Desktop\\train_update.csv')

# Read the first chunk to identify categorical columns
first_chunk = pd.read_csv(file_path, dtype=dtypes, chunksize=chunk_size).__next__()
categorical_columns = [col for col in first_chunk.columns if first_chunk[col].dtype not in numerics and col != 'MachineIdentifier' and col != first_chunk.columns[-1]]

# Generate mappings for each categorical column
all_mappings = {}
for column_name in categorical_columns:
    unique_values = pd.read_csv(file_path, usecols=[column_name], dtype=dtypes)[column_name].unique()
    all_mappings[column_name] = {value: code for code, value in enumerate(unique_values)}

# Initialize an empty DataFrame to store processed chunks
processed_chunks = []

# Read and process the file in chunks
for chunk in pd.read_csv(file_path, dtype=dtypes, chunksize=chunk_size):
    for column_name in categorical_columns:
        chunk[column_name] = chunk[column_name].map(all_mappings[column_name])
    processed_chunks.append(chunk)

# Concatenate all processed chunks into a single DataFrame
train = pd.concat(processed_chunks)

# Save the processed DataFrame to a CSV file
output_csv_path = 'C:\\Users\\rp7248\\Desktop\\new_data.csv'
train.to_csv(output_csv_path, index=False)

# Write mappings to a text file
output_file_path = 'C:\\Users\\rp7248\\Desktop\\mapping.txt'
with open(output_file_path, 'w') as file:
    for column_name, mapping in all_mappings.items():
        formatted_mappings = [f"{value}: {code}" for value, code in mapping.items()]
        mapping_array_str = column_name + " = " + "[" + ", ".join(formatted_mappings) + "]" + "\n\n"
        file.write(mapping_array_str)
```
